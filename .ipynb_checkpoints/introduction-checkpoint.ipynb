```python
def factorial(x):
    
    if x == 1:
        return 1
    else:
        return (x * factorial(x-1))

```

```python
factorial(5)
```

```python
def compare_to_five(y):
    if y > 5:
        return "Greater"
    elif y < 5:
        return "Less"
    else:
        return "Equal"
```
